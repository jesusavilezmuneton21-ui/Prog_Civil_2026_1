3. Sí $ITH > 78 \space and \space ITH <= 83$ peligro por estrés térmico
4. Sí $ITH > 83$ situación crítica

## Teoría del Confort Térmico en Ganado Vacuno

El confort térmico en el ganado vacuno se refiere a un estado en el que el animal no experimenta estrés por frío ni por calor. Mantener a los animales dentro de su zona de confort térmico es crucial para su salud, bienestar y productividad (producción de leche, carne, reproducción).

### Estrés por Calor
El estrés por calor es una preocupación significativa en muchas regiones, especialmente en ambientes cálidos y húmedos. Ocurre cuando la carga de calor total del animal excede su capacidad para disipar el calor. Las principales fuentes de calor para el ganado incluyen:

1.  **Metabolismo:** Producción interna de calor como resultado de procesos fisiológicos.
2.  **Radiación solar:** Absorción de energía radiante del sol.
3.  **Convección:** Transferencia de calor desde el aire circundante si la temperatura ambiente es superior a la temperatura de la piel del animal.
4.  **Conducción:** Transferencia de calor por contacto directo con superficies más calientes.

Cuando el ganado experimenta estrés por calor, muestra una serie de respuestas fisiológicas y de comportamiento para intentar reducir su carga térmica:

*   **Aumento de la frecuencia respiratoria (jadeo):** Es el mecanismo más importante para la disipación de calor por evaporación.
*   **Reducción del consumo de alimento:** Para disminuir la producción de calor metabólico.
*   **Aumento del consumo de agua:** Para reponer las pérdidas por evaporación y mantener la hidratación.
*   **Búsqueda de sombra o zonas frescas:** Cambios de comportamiento para evitar la radiación solar directa.
*   **Disminución de la producción:** La producción de leche y carne, así como la eficiencia reproductiva, se ven afectadas negativamente.

### Índice de Temperatura y Humedad (ITH)
El ITH es uno de los índices más utilizados para evaluar el nivel de estrés por calor en el ganado. Combina la temperatura del aire y la humedad relativa, ya que ambos factores influyen en la capacidad del animal para disipar calor por evaporación. Un ITH alto indica una mayor probabilidad de estrés por calor.

La fórmula general para el ITH y sus interpretaciones varían ligeramente, pero el principio es el mismo: **a medida que el ITH aumenta, el riesgo de estrés térmico y sus consecuencias negativas para el ganado también aumentan.** Los umbrales de ITH para las diferentes categorías de estrés (no estresante, alerta, peligro, crítico) proporcionan una guía para implementar medidas de mitigación.

### Estrategias de Mitigación
Las estrategias para reducir el estrés por calor incluyen:

*   **Sombra:** Proporcionar acceso a sombra natural o artificial.
*   **Ventilación:** Mejorar el flujo de aire en las instalaciones.
*   **Aspersores y ventiladores:** Utilizar sistemas de enfriamiento evaporativo.
*   **Acceso a agua fresca:** Asegurar disponibilidad constante de agua de calidad.
*   **Manejo nutricional:** Ajustar la dieta para reducir la producción de calor metabólico.
*   **Manejo del pastoreo:** Evitar el pastoreo durante las horas más calurosas del día.

```python
# ENTRADA: pedir tbs (temperatura de bubo seco) y tbh (temperatura de ulbo humedo)
tbs = float(input("Ingrese la temperatura bulbo seco, grados centigrados: "))
tbh = float(input("Ingrese la temperatura bulbo húmedo, grados centigrados: "))

# PROCESO: calculo de ith (indice de temperatura y humedad)
ith = 0.72 * (tbs + tbh) + 40.6

# SALIDA: imprimir resultado, de la evaluacion del ith y el control termico
if ith <= 70:
  print("ambiente no estresante")
elif ith > 70 and ith <= 78:
  print("situacion de alerta por estrés térmico")
elif ith > 78 and ith <= 83:
  print("peligro por estrés térmico")
elif ith > 83:
  print("la situacion es critica")
```
