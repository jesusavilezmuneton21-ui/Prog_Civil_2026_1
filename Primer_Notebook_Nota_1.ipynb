# Problema de ingeniería (nivel básico)

Un estudiante de ingeniería necesita **calcular el área de un terreno rectangular** para estimar el espacio de cultivo.

La fórmula es:

[
Área = base \times altura
]

---

# 1. Definir el algoritmo (Entradas – Procesos – Salidas)

### Entradas

Datos que el programa debe recibir:

* base del terreno (metros)
* altura del terreno (metros)

### Procesos

Operaciones que realizará el algoritmo:

* multiplicar base × altura

### Salidas

Resultado que mostrará el programa:

* área del terreno en **m²**

---

# 2. Algoritmo (descripción paso a paso)

1. Iniciar programa
2. Pedir al usuario la **base del terreno**
3. Pedir al usuario la **altura del terreno**
4. Calcular el área usando la fórmula
5. Mostrar el área del terreno
6. Finalizar programa

---

# 3. Pseudocódigo

```
Inicio

Leer base
Leer altura

area = base * altura

Mostrar "El área del terreno es:", area

Fin
```

---

# 4. Código en Python

```python
# Programa para calcular el área de un terreno rectangular

# Entrada de datos
base = float(input("Ingrese la base del terreno en metros: "))
altura = float(input("Ingrese la altura del terreno en metros: "))

# Proceso (cálculo)
area = base * altura

# Salida
print("El área del terreno es:", area, "m²")
```

---

# 5. Ejemplo de ejecución

```
Ingrese la base del terreno en metros: 10
Ingrese la altura del terreno en metros: 5

El área del terreno es: 50 m²
```
