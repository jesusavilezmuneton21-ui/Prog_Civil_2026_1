#**Dosificación de mezcla para columnas**

**Planteamiento**

 Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

  Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

Requerimientos:

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$v = \pi * r^2 * h$$
 (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos:
$$sacos = v * 9.5$$

**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58
 (el estudiante verá aquí la importancia de los tipos decimales).


```python
 #ENTRADA: Solicitar la altura de la columna en metros.
 # Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

#PROCESO: calcula el radio r, calcula el volumen de la columna, calcula sacos
r = d / 2
v = 3.1416 * r**2 * h
sacos = v * 9.5

#SALIDA: Imprime el resultado
print("Volumen ≈", round(v, 3), "m3")
print("Sacos necesarios ≈", round(sacos, 2))
```

El código realiza los siguientes pasos:

*   `h = float(input("Ingrese la altura de la columna en metros: "))`: Solicita al usuario la altura de la columna en metros y la guarda como un número decimal en la variable `h`.

*   `d = float(input("Ingrese el diámetro de la columna en metros: "))`: Solicita al usuario el diámetro de la columna en metros y la guarda como un número decimal en la variable `d`.

*   `r = d / 2`: Calcula el radio de la columna dividiendo el diámetro `d` entre 2 y almacena el resultado en la variable `r`.

*   `v = 3.1416 * r**2 * h`: Calcula el volumen de la columna cilíndrica utilizando la fórmula `π * r² * h`, donde `3.1416` es una aproximación de Pi, `r**2` es el radio al cuadrado, y `h` es la altura. El resultado se guarda en la variable `v`.

*   `sacos = v * 9.5`: Calcula la cantidad total de sacos de cemento multiplicando el volumen de la columna `v` por la cantidad de sacos necesarios por metro cúbico (`9.5`). El resultado se almacena en la variable `sacos`.

*   `print("Volumen ≈", round(v, 3), "m3")`: Imprime el volumen calculado `v`, redondeado a tres decimales, seguido de la unidad `m3`.

*   `print("Sacos necesarios ≈", round(sacos, 2))`: Imprime la cantidad de sacos de cemento `sacos` necesaria, redondeada a dos decimales.
